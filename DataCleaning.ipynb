```python
import pandas as pd
import numpy as np
```

```python
COWcode = pd.read_csv("COW-country-codes.csv")
COWcode = COWcode.drop_duplicates()
COWcode.to_csv("COW-country-codes.csv", index=False)
COWcode = pd.read_csv("COW-country-codes.csv")
COWcode
```

```python
votes = pd.read_csv("UNVotes-UTF-8.csv", dtype={"short":str})
```

```python
print(votes.columns.values)
votes
```

```python
votes = votes.drop(["Unnamed: 0","short","descr","ident"],axis=1)
print(votes.columns.values)
votes
```

```python
votes = votes[votes["ccode"] != 345] ## exclude ccode 345 (Yugoslavia) because it is not unique in each vote (rcid)
print(votes.columns.values)
votes
```

```python
votes = votes.drop("year",axis=1)
votes["dateNew"] = pd.to_datetime(votes["date"], format="%Y-%m-%d")
votes["year"] = votes["dateNew"].dt.year
print(votes.columns.values)
votes
```

```python
votes = votes.rename({"Countryname":"CountrynameOld"}, axis=1)
votes["Countryname"] = votes["Country"].where(pd.isna(votes["CountrynameOld"]),votes["CountrynameOld"])
votes = votes.drop("CountrynameOld",axis=1)
print(votes.columns.values)
votes
```

```python
votes = votes.merge(COWcode, how="left", left_on="ccode", right_on="CCode")
print(votes.columns.values)
votes
```

```python
votes = votes.drop(["Country","CCode"],axis=1)
votes = votes.rename({"StateAbb":"Country","StateNme":"CountrynameCOW"}, axis=1)
print(votes.columns.values)
votes
```

```python
colOrder = ['date', 'dateNew', 'year',
            'rcid', 'resid', 'session',
            'ccode', 'Country', 'Countryname', 'CountrynameCOW', 'vote',
            'unres', 'yes', 'no', 'abstain', 'amend', 'para', 'importantvote',
            'me', 'nu', 'di', 'hr', 'co', 'ec']
votes = votes[colOrder]
print(votes.columns.values)
votes
```

```python
votes = votes.sort_values(by = ["dateNew","resid","ccode"])
print(votes.columns.values)
votes
```

```python
votes = votes[votes["year"]>=1991] ## 1991 to 2021
votes = votes.drop("dateNew",axis=1)
print(votes.columns.values)
votes
```
